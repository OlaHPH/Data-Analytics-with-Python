```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

import yfinance as yf
```

```python
#downloading historical stock prices of Marimekko and Lindex since 01.01.2017 from Yahoo Finance services
#Marimekko and Lindex are 2 Finnish companies operating in the same sector - Consumer cyclical.
#Marimekko's industry is apparel manufacturing while Lindex's is apparel retail.
mekko = yf.download('MEKKO.HE', start='2017-01-01')
lindex = yf.download('LINDEX.HE', start='2017-01-01')
```

Stock quote brief explanation
1. Close - closing price is the last trading price during the regular trading hours on a given day.
Example: Helsinki Stock Market's regular trading hours are 10:00-18:30, with the exception of Holiday.
Usually previous closing price (Close) would be the next session's opening price (Open), but not always.
2. High- highest traded price during the given day.
3. Low - lowest traded price during the given day.
4. Open - first trade price that was recorded during the dayâ€™s trading.
5. Volume - number of stocks traded in a given period.  

```python
#Plotting historical closing prices of Marimekko and Lindex stocks over time
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12,5))
#marimekko
mekko['Close'].plot(ax=axs[0],color='black')
axs[0].set_title("Marimekko' stocks historical price since 2017")
axs[0].set_ylabel('Euros', loc='top')
#lindex
lindex['Close'].plot(ax=axs[1],color='red')
axs[1].set_title("Lindex' stocks historical price since 2017")
axs[1].set_ylabel('Euros', loc='top')
```

```python
#plotting the 2 in 1 field
fig, ax = plt.subplots()
mekko['Close'].plot(ax=ax,color='black')
lindex['Close'].plot(ax=ax,color= 'red')
plt.ylabel('Euros', loc='top')
plt.title('Marimekko and Lindex historical prices since 2017')
```

General observations.
1. Marimekko has a higher trading price in comparison to Lindex at the moment.
2. Both stocks have been fluctuating over the years.
3. Marimekko' stock price has risen from approx. under 2 euros to approx. above 14 euros in the given period (01.01.2017-21.02.2025)
4. Lindex' stock price has dropped from approx. 6 euros to approx. under 3 euros in the given period (01.01.2017-21.02.2025)

###Agregation

```python
# Using resample() to get the monthly average closing prices from the given daily prices
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12,5))
#marimekko
mekko['Close'].resample('ME').mean().plot(ax=axs[0],color='black')
axs[0].set_title('Marimekko monthly average closing prices since 2017')
axs[0].set_ylabel('Euros', loc='top')
#lindex
lindex['Close'].resample('ME').mean().plot(ax=axs[1],color='red')
axs[1].set_title('Lindex monthly average closing prices since 2017')
axs[1].set_ylabel('Euros', loc='top')
```

```python
#Average quarterly closing price
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12,5))
#marimekko
mekko['Close'].resample('QE').mean().plot(ax=axs[0],color='black')
axs[0].set_title('Marimekko quarterly average closing prices since 2017')
axs[0].set_ylabel('Euros', loc='top')
#lindex
lindex['Close'].resample('QE').mean().plot(ax=axs[1],color='red')
axs[1].set_title('Lindex quarterly average closing prices since 2017')
axs[1].set_ylabel('Euros', loc='top')
```

```python
#Annual average closing price
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12,5))
#marimekko
mekko['Close'].resample('YE').mean().plot(ax=axs[0],color='black')
axs[0].set_title('Marimekko annual average closing prices since 2017')
axs[0].set_ylabel('Euros', loc='top')
#lindex
lindex['Close'].resample('YE').mean().plot(ax=axs[1],color='red')
axs[1].set_title('Lindex annual average closing prices since 2017')
axs[1].set_ylabel('Euros', loc='top')
```

We can notice a noticeable fall in in Lindex average closing prices in 2020-2021.
Citing info from Lindex Group: Lindex Group has been systematically implementing the restructuring programme which was approved in **February 2021**.
Source: https://www.lindex-group.com/en/investors/stockmann-as-an-investment/

```python
# quarterly volumes
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12,5))
#marimekko
mekko['Volume'].resample('QE').sum().plot(ax=axs[0],color='black')
axs[0].set_title('Marimekko quarterly volumes since 2017')
axs[0].set_ylabel('10 millions of shares')
#lindex
lindex['Volume'].resample('QE').sum().plot(ax=axs[1],color='red')
axs[1].set_title('Lindex quarterly volumes since 2017')
axs[1].set_ylabel('10 millions of shares')
```

```python
# yearly volumes
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12,5))
#marimekko
mekko['Volume'].resample('YE').sum().plot(ax=axs[0],color='black')
axs[0].set_title('Marimekko yearly volumes since 2017')
axs[0].set_ylabel('10 millions of shares')
#lindex
lindex['Volume'].resample('YE').sum().plot(ax=axs[1],color='red')
axs[1].set_title('Lindex yearly volumes since 2017')
axs[1].set_ylabel('10 millions of shares')
```

###Moving averages

 Moving averages, i.e the average price that traders pay for an asset or security over a specific period,is one of the common methods used in technical analysis. The most common moving averages are 50 (for short-term patterns), 100, and 200 days (for long-term trends).

```python
# moving averages - 50 and 200 days in comparision with original values
fig, ax = plt.subplots()

mekko['Close'].plot(ax=ax,title='Marimekko stocks moving averages')
mekko['Close'].rolling(50).mean().plot(ax=ax)
mekko['Close'].rolling(200).mean().plot(ax=ax)
ax.set_ylabel('Euros', loc='top')

plt.legend(['Closing prices' , 'Moving average (50 days)' ,'Moving average (200 days)'])
```

```python
# moving averages - 50 and 200 days in comparision with original values
fig, ax = plt.subplots()

lindex['Close'].plot(ax=ax, title ='Lindex stock moving averages')
lindex['Close'].rolling(50).mean().plot(ax=ax)
lindex['Close'].rolling(200).mean().plot(ax=ax)
ax.set_ylabel('Euros', loc='top')
plt.legend(['Closing prices' , 'Moving average (50 days)','Moving average (200 days)'])
```

### Percentage of changes

```python
# percentage changes for closing prices
mekko['mekko_change%'] = (mekko['Close'].pct_change())*100
lindex['lindex_change%'] = (lindex['Close'].pct_change())*100
```

```python
changes =pd.concat([mekko['mekko_change%'], lindex['lindex_change%']],axis=1).fillna(0)
changes
```

```python
(changes.loc['2024-12-01':'2024-12-31']).plot(color={'black','red'})
plt.axhline(color='green')
plt.title('Marimekko and Lindex percentages of changes in 2024')
```

```python
(changes[-30:]*100).plot.bar(figsize=(12,5),color={'black','red'})
plt.ylabel('Daily percentage of change')
plt.title('Last 30 daily changes of Marimekko and Lindex stocks ')
plt.axhline(color='green')
```

```python
#statistical numbers for change%
(changes*100).describe().round(2)
```

```python
sns.boxplot(data=changes*100, showmeans='True')
```

```python
#big changes where the changes is bigger than 10%
changes[(abs(changes['mekko_change%'])>10) | (abs(changes['lindex_change%'])>10)]
```

###Moving volatility

Volatility is a statistical measure of the dispersion i.e. the changes of returns for a given security or market index.The bigger the volatility, the higher the risk.

```python
#moving volatility of 100 days
plt.figure(figsize=(12,5))

# Marimekko stock
(changes['mekko_change%'].rolling(100).std() * (252**0.5)).plot(label = 'Mekko', legend = True, color ='black')

# Lindex stock
(changes['lindex_change%'].rolling(100).std() * (252**0.5)).plot(label = 'Lindex', legend = True, color ='red')
```

```python
#moving volatility of 252 days (instead of 365 days as 252 are the trading days over the year)
plt.figure(figsize=(12,5))

# Marimekko stock
(changes['mekko_change%'].rolling(252).std() * (252**0.5)).plot(label = 'Mekko', legend = True, color = 'black')

# Lindex stock
(changes['lindex_change%'].rolling(252).std() * (252**0.5)).plot(label = 'Lindex', legend = True, color = 'red')
```
